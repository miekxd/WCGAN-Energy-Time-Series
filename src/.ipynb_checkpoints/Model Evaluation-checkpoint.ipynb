## Plot Training and Test Losses

```python
model='256-30000-1464-1464-720-0.00012-0.00014'

data=[]

with open(f"/home/ec2-user/SageMaker/WCGAN-DR/output/{model}/losses.txt", "r") as f:
  # Iterate through each line
  for line in f:
    items = line.strip().split(',')

    # Convert extracted values to numbers (if necessary)
    epoch = float(items[0].split(': ')[-1])  # Extract epoch value (assuming format)
    rmse = float(items[1].split(': ')[-1])  # Extract RMSE value
    test_rmse = float(items[2].split(': ')[-1])  # Extract Test RMSE value

    # Create a dictionary for each line's data
    data_dict = {"Epoch": epoch, "RMSE": rmse, "Test RMSE": test_rmse}
    data.append(data_dict)

epochs = [item["Epoch"] for item in data]
rmse = [item["RMSE"] for item in data]
test_rmse = [item["Test RMSE"] for item in data]

threshold=1

rmse = [min(val, threshold) for val in rmse]
test_rmse = [min(val, threshold) for val in test_rmse]


# Create the plot
plt.figure(figsize=(10, 6))

# Plot RMSE
plt.plot(epochs, rmse, label='RMSE')

# Plot Test RMSE
plt.plot(epochs, test_rmse, label='Test RMSE')

# Set labels and title
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.title('RMSE and Test RMSE vs Epochs')

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Evaluate Model Performance

### Performance on December Energy Data


```python
from dense3_skip2 import Generator
from data import Data
from main import load_checkpoint
import torch
import numpy as np
import pandas as pd
import math
from numpy import zeros, ones, expand_dims
from numpy.random import randn, randint
import matplotlib.pyplot as plt

model='256-100000-1464-1464-720-6e-05-0.0001'#name of the training job

path ='/home/ec2-user/SageMaker/WCGAN-DR/output/'+model+ '/gen.pth'

#parse the string into variables
hyper=model.split('-')
latent_dim=int(hyper[2])
ts_dim=int(hyper[3])
conditional=int(hyper[4])

n=4 #number of plots you want to make (square root of n must be an integer) 
arrayno=0

device = torch.device('cpu')
G = Generator(latent_dim, ts_dim, conditional)
G.load_state_dict(torch.load(path, map_location=device))

data=Data(ts_dim,"/home/ec2-user/SageMaker/WCGAN-DR/data/EnergyData/DecEnergyData.csv")
fake_lines, real_lines, start_prices = data.get_samples(G=G, latent_dim=latent_dim, n=n, ts_dim=ts_dim,conditional=conditional, use_cuda=False)

real_lines = np.squeeze(real_lines.cpu().data.numpy())
fake_lines = np.squeeze(fake_lines.cpu().data.numpy())

real = np.array([data.post_processing_specific(arrayno,real_lines[i], start_prices[i]) for i in range(real_lines.shape[0])])
fake = np.array([data.post_processing_specific(arrayno,fake_lines[i], start_prices[i]) for i in range(real_lines.shape[0])])
real = np.squeeze(real)
fake = np.squeeze(fake)

real=real[:,conditional:]
fake=fake[:,conditional:]


fig, ax = plt.subplots(int(math.sqrt(n)),int(math.sqrt(n)),figsize=(10, 10))
ax = ax.flatten()
fig.suptitle('Energy Prediction')
for i in range(ax.shape[0]):
    ax[i].plot(real[i], color='red', label='Real', alpha =0.7)
    ax[i].plot(fake[i], color='green', label='Fake', alpha =0.7)
    ax[i].legend()
```

## Generate fake dataset and plot against real dataset


```python
from dense3_skip2 import Generator
from data import Data
import torch
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
    
model='256-100000-1464-1464-720-6e-05-0.0001'

path ='/home/ec2-user/SageMaker/WCGAN-DR/output/'+model+ '/gen.pth'

hyper=model.split('-')
latent_dim=int(hyper[2])
ts_dim=int(hyper[3])
conditional=int(hyper[4])

idx=168

device = torch.device('cpu')
G = Generator(latent_dim, ts_dim, conditional)
G.load_state_dict(torch.load(path, map_location=device))

data=Data(ts_dim,"/home/ec2-user/SageMaker/WCGAN-DR/data/EnergyData/test.csv")

fake_energy, fake_temp, real_energy, real_temp = data.generate_data(idx=idx, G=G, latent_dim=latent_dim, ts_dim=ts_dim, conditional=conditional, use_cuda=False)
    

fake_energy = fake_energy[conditional:]
fake_temp = fake_temp[conditional:]
real_energy = real_energy[conditional:]
real_temp = real_temp[conditional:]


slope = (np.max(fake_temp)-np.max(real_temp))/(ts_dim-conditional)

for i in range(fake_temp.shape[0]):
    fake_temp[i]=fake_temp[i]-slope*i
print(fake_temp.shape)

data_graph = pd.DataFrame({'hours': np.arange(0, 744, step=1),
                          'energy usage (kWh)': fake_energy, 
                          'outside temp (C)': fake_temp,
                          'real temp': real_temp,
                          'real energy': real_energy})

data = pd.DataFrame({'hours': np.arange(0, 744, step=1),
                          'energy usage (kWh)': fake_energy, 
                          'outside temp (C)': fake_temp})

import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 6)) 

# Plot energy data on the first subplot (ax1)
ax1.plot(data_graph['hours'], data_graph['energy usage (kWh)'], label='Fake Energy', color='g')  # Blue for fake energy
ax1.plot(data_graph['hours'], data_graph['real energy'], label='Real Energy', color='r')  # Green for real energy
ax1.set_xlabel('Time (hourly)')
ax1.set_ylabel('kWh')
ax1.set_title('Energy Consumption')
ax1.legend()

# Plot temperature data on the second subplot (ax2)
ax2.plot(data_graph['hours'], data_graph['outside temp (C)'], label='Fake Temp', color='g')  # Blue for fake temp
ax2.plot(data_graph['hours'], data_graph['real temp'], label='Real Temp', color='r')  # Green for real temp
ax2.set_xlabel('Time (hourly)')
ax2.set_ylabel('Temperature (C)')
ax2.set_title('Temperature')
ax2.legend()

# Adjust layout (optional)
plt.tight_layout()  # Reduces spacing between subplots

plt.show()



data_graph.head(744)
```

### Flexible rules

```python
import openpyxl

path='/home/ec2-user/SageMaker/WCGAN-DR/fakedata/fake_data.xlsx'
# Load the Excel workbook
workbook = openpyxl.load_workbook(path)

# Select the appropriate worksheet
worksheet = workbook['Sheet1']  # Change 'Sheet1' to your sheet name

# Add a header for the Demand Response column
worksheet['D1'] = 'demand response'

afe=np.mean(fake_energy)
afe=150
mfe=np.max(fake_energy)
aft=np.mean(fake_temp)
aft=25


# Iterate through the rows starting from the second row
for row_number in range(2, worksheet.max_row + 1):
    # Get the values of Energy and Temperature from columns B and C
    energy = worksheet[f'B{row_number}'].value
    temperature = worksheet[f'C{row_number}'].value
    
    # Determine the Demand Response based on energy usage and temperature
    if energy < 100:
        if temperature < 22:
            response = temperature  # Maintain
        elif 22 <= temperature <= 26:
            response = temperature  # Maintain
        else:
            response = temperature-(temperature-aft)*((mfe-energy)/mfe)
            if response >= 30:
                response=29 # Lower AC temp
    elif 100 <= energy <= 200:
        if temperature < 22:
            response = temperature  # Maintain
        elif 22 <= temperature <= 26:
            response = temperature # Maintain
        else:
            response = temperature-(temperature-aft)*((mfe-energy)/mfe)  # Lower AC temp
    else:  # Energy > 200
        if temperature < 22:
            response = temperature-(temperature-aft)*((mfe-energy)/mfe)  # Raise AC temp
        elif 22 <= temperature <= 26:
            response = temperature  # Maintain
        else:
            response = temperature  # Maintain
    
    response = round(response*2)/2
    # Write the computed Demand Response to the corresponding cell in column D
    worksheet[f'D{row_number}'] = response
    
# Save the changes
workbook.save(path) 
```

### Peg temperature setting to ideal temperature

```python
import openpyxl

path='/home/ec2-user/SageMaker/WCGAN-DR/fakedata/fake_data.xlsx'
# Load the Excel workbook
workbook = openpyxl.load_workbook(path)

# Select the appropriate worksheet
worksheet = workbook['Sheet1']  # Change 'Sheet1' to your sheet name

# Add a header for the Demand Response column
worksheet['D1'] = 'demand response'

afe=np.mean(fake_energy)
afe=75 #ideal energy usage
mfe=np.max(fake_energy)

aft=np.mean(fake_temp)
aft=25 #ideal temperature


# Iterate through the rows starting from the second row
for row_number in range(2, worksheet.max_row + 1):
    # Get the values of Energy and Temperature from columns B and C
    energy = worksheet[f'B{row_number}'].value
    temperature = worksheet[f'C{row_number}'].value
    
    response = temperature-(temperature-aft)*((mfe-energy)/mfe)
    response = round(response*2)/2
    # Write the computed Demand Response to the corresponding cell in column D
    worksheet[f'D{row_number}'] = response
    
    
# Save the changes
workbook.save(path) 
```
